```python
import pandas as pd
```

```python
ev=pd.read_csv('EVcharging.csv')
ev
```

```python
ev.columns
```

```python
ev = ev.rename(columns={
    'Usage Stats (avg users/day)': 'demand',
    'Charging Capacity (kW)': 'capacity',
    'Cost (USD/kWh)': 'cost',
    'Distance to City (km)': 'distance'
})
```

```python
ev
```

```python
X = ev[['capacity','cost','distance','Parking Spots','Reviews (Rating)']]
y = ev['demand']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
ev['predicted_demand'] = model.predict(X)
```

```python
ev['predicted_demand']
```

```python
ev.to_csv("ev_output.csv", index=False)
```
