### DATA STANDARDIZATION

```python
us['country']='US'
in_you['Country']='India'
ca['country']='Canada'
```

```python
data=pd.concat([us,in_you,ca],ignore_index=True)
```

```python
data['trending_date']=pd.to_datetime(data['trending_date'],format='%y.%d.%m')
data['publish_time']=pd.to_datetime(data['publish_time'])
```

### Remove Duplicates

```python
data.drop_duplicates(inplace=True)
```

```python
data
```

```python
data['tags'] = data['tags'].fillna('No Tags')
```

```python
data
```

### FEATURE ENGINEERING

```python
data['engagement_rate']=(data['likes']+data['comment_count'])/data['views']
```

### TRENDING DURATION

```python
data['trending_duration']=(data.groupby('video_id')['trending_date'].transform(lambda x:(x.max() - x.min()).days))
```

```python
data
```

```python
data["publish_time"] = pd.to_datetime(data["publish_time"]).dt.tz_localize(None)
data["trending_date"] = pd.to_datetime(data["trending_date"], format="%y.%d.%m")
```

```python
from textblob import TextBlob

data["sentiment_score"] = data["title"].apply(
    lambda x: TextBlob(str(x)).sentiment.polarity
)

def sentiment_label(score):
    if score > 0:
        return "Positive"
    elif score < 0:
        return "Negative"
    else:
        return "Neutral"

data["sentiment"] = data["sentiment_score"].apply(sentiment_label)
```
